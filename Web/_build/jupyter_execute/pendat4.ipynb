# **FUZZY C-MEANS**

## FUZZY C-MEANS CLUSTERING

K-Means Clustering merupakan metode pengelompokan data yang bertujuan memisahkan data ke dalam beberapa kelompok (cluster) berdasarkan kemiripan fitur. Data dengan karakteristik serupa akan ditempatkan dalam satu kelompok, sementara data yang berbeda akan dikelompokkan ke dalam klaster lain. Proses pengelompokan ini menggunakan centroid (titik pusat) sebagai referensi dalam menentukan keanggotaan tiap data pada suatu klaster. Tujuannya adalah untuk memastikan bahwa data yang memiliki kesamaan berada dalam kelompok yang sama, dan data yang berbeda tersebar di kelompok yang berbeda.


Tidak seperti K-Means yang menerapkan pendekatan hard clustering (setiap data hanya masuk ke satu cluster secara mutlak), Fuzzy C-Means (FCM) menggunakan pendekatan soft clustering, di mana setiap data dapat memiliki tingkat keanggotaan pada lebih dari satu cluster. Pendekatan ini lebih sesuai untuk data yang memiliki batas antar cluster yang tidak jelas atau bersifat ambigu.

### KONSEP DASAR FUZZY C-MEANS CLUSTERING
Fuzzy C-Means (FCM) Clustering memiliki konsep dasar di mana setiap data tidak langsung dikategorikan ke dalam satu cluster tertentu secara mutlak, melainkan diberikan nilai keanggotaan (membership value) yang menunjukkan seberapa besar kemungkinan data tersebut termasuk dalam masing-masing cluster. Nilai ini berada dalam rentang 0 hingga 1, dan untuk setiap data, total dari semua nilai keanggotaannya terhadap seluruh cluster harus sama dengan 1.

- Beberapa elemen kunci dalam FCM antara lain:

- Titik Data: Kumpulan objek yang akan dikelompokkan.

- Centroid: Titik representatif atau pusat dari masing-masing cluster.

- Matriks Keanggotaan Fuzzy (U): Matriks yang menyimpan tingkat keanggotaan setiap data terhadap seluruh cluster.

- Parameter Keburaman (m): Faktor yang menentukan tingkat ketidakpastian dalam penentuan keanggotaan (umumnya

### ALGORITMA FUZZY C-MEANS CLUSTERING

Proses kerja algoritma Fuzzy C-Means adalah sebagai berikut:

1. Inisialisasi:
    
- Tentukan jumlah cluster
ùê∂ dan nilai fuzziness ùëö > 1 m>1 (biasanya 2).

- Inisialisasi matriks keanggotaan
ùëà=[Uùëñùëó ] (membership matrix* $U_{ij}$)secara acak dengan syarat:
   
     $$
     \sum_{j=1}^c U_{ij} = 1, \quad \forall i
     $$


2. Hitung Centroid Tiap Cluster:

   $$
   v_j = \frac{\sum_{i=1}^n (U_{ij})^m \cdot x_i}{\sum_{i=1}^n (U_{ij})^m}
   $$

   Dengan keterangan sebagai berikut:

- vj menyatakan posisi pusat (centroid) dari cluster ke-ùëó
- ùë•ùëñ merupakan data atau objek ke-ùëñ dalam himpunan data
- ùëàùëñùëó adalah derajat keanggotaan dari data ke-ùëñ terhadap cluster ke-ùëó


3. Update Derajat Keanggotaan:

   $$
   U_{ij} = \frac{1}{\sum_{k=1}^c \left( \frac{\|x_i - v_j\|}{\|x_i - v_k\|} \right)^{\frac{2}{m-1}}}
   $$

- Jika ùë•ùëñ = ùëêùëó, maka set ùë¢ùëñùëó = 1 dan lainnya 0 (untuk hindari pembagian 0).


4. Periksa Konvergensi:

- Jika perubahan U atau Cij antar iterasi < epsilon (nilai kecil), maka berhenti.

## IMPLEMENTASI PYTHON

Berikut juga saya lakukan visualisasi 3 Dimensi pada data Iris setelah clustering

```python
import numpy as np
import pandas as pd

# Data dan parameter
X = np.array([1, 2, 4, 5, 8, 9], dtype=float)
c = 2          # jumlah cluster
m = 2.0        # fuzziness
epsilon = 0.01  # kriteria konvergensi
n = len(X)
max_iter = 100  # batas maksimal iterasi

# Inisialisasi matriks U secara manual (2x6)
U = np.array([
    [0.8, 0.6, 0.5, 0.4, 0.2, 0.1],
    [0.2, 0.4, 0.5, 0.6, 0.8, 0.9]
])

def calc_centroids(U, X, m):
    um = U ** m
    v = (um @ X) / np.sum(um, axis=1)
    return v

def update_U(X, v, m):
    c = len(v)
    n = len(X)
    U_new = np.zeros((c, n))
    for i in range(n):
        for j in range(c):
            d_ji = abs(X[i] - v[j])
            if d_ji == 0:
                U_new[:, i] = 0
                U_new[j, i] = 1
            else:
                denom = sum([(d_ji / max(abs(X[i] - v[k]), 1e-10))**(2 / (m - 1)) for k in range(c)])
                U_new[j, i] = 1 / denom
    return U_new

def objective_function(U, X, v, m):
    J = 0
    c, n = U.shape
    for j in range(c):
        for i in range(n):
            J += (U[j, i] ** m) * (abs(X[i] - v[j]) ** 2)
    return J

# Simpan histori
history = []
for iteration in range(max_iter):
    v = calc_centroids(U, X, m)
    J = objective_function(U, X, v, m)
    history.append((iteration + 1, v.copy(), U.copy(), J))

    U_new = update_U(X, v, m)
    v_new = calc_centroids(U_new, X, m)

    # Konvergensi jika perbedaan semua centroid < epsilon
    if np.all(np.abs(v_new - v) < epsilon):
        break
    U = U_new

# Konversi ke DataFrame untuk tampilan rapi
results = []
for iter_num, centroids, membership, obj_val in history:
    result = {
        "Iterasi": iter_num,
        "Centroid 1": centroids[0],
        "Centroid 2": centroids[1],
        "Objektif": obj_val
    }
    for j in range(c):
        for i in range(n):
            result[f"U[{j+1},{i+1}]"] = membership[j, i]
    results.append(result)

df_results = pd.DataFrame(results)
print(df_results.round(4))  # pembulatan 4 desimal
```

## Clustering menggunakan Fuzzy C-Means

### Clustering dengan 2 cluster dan Visualisasi data

Disini Saya menggunakan algoritma Fuzzy C-Means (FCM) untuk membagi data menjadi dua klaster. Sebelumnya, data dinormalisasi agar setiap fitur memiliki skala yang sama, lalu ditranspos agar sesuai dengan format input dari FCM. Algoritma ini berjalan dengan parameter standar, dan menghasilkan pusat klaster, nilai keanggotaan tiap data ke masing-masing klaster, serta nilai fungsi objektif di setiap iterasi untuk memantau proses konvergensinya.

Setelah proses klasterisasi selesai, saya mengevaluasi hasilnya menggunakan dua metrik: Silhouette Score untuk melihat seberapa baik pemisahan antar klaster, dan Fuzzy Partition Coefficient (FPC) untuk mengukur kejelasan keanggotaan data. Hasil akhirnya juga divisualisasikan dalam bentuk scatter plot 2 dimensi menggunakan PCA, sehingga kita bisa melihat secara langsung bagaimana data terbagi dalam dua klaster yang berbeda.

```python

# Transpose fitur agar sesuai dengan input skfuzzy (fitur x sample)
data = scaled_features.T

# Fuzzy C-Means clustering
n_clusters = 2
cntr, u, u0, d, jm, p, fpc = fuzz.cluster.cmeans(
    data, c=n_clusters, m=2.0, error=0.005, maxiter=1000, init=None, seed=42
)

# Ambil label keras dari keanggotaan tertinggi
labels_2 = np.argmax(u, axis=0)

# Evaluasi clustering
silhouette_fuzzy = silhouette_score(scaled_features, labels_2)

# Cetak hasil evaluasi
print(f"[Fuzzy C-Means - 2 Clusters]")
print(f"Silhouette Score: {silhouette_fuzzy:.4f}")
print(f"Fuzzy Partition Coefficient (FPC): {fpc:.4f}")
print(f"Objective Function Value (Jm): {jm[-1]:.4f}")  # ambil nilai terakhir iterasi
print(f"Cluster Centers (in scaled feature space):\n{cntr}")

# Menampilkan nilai fungsi objektif di setiap iterasi
print("Objective Function Values per Iteration:")
for i, val in enumerate(jm):
    print(f"Iterasi {i+1:02d}: {val:.4f}")


# Visualisasi hasil clustering (dalam ruang PCA)
plt.figure(figsize=(6, 4))
plt.scatter(reduced_features[:, 0], reduced_features[:, 1], c=labels_2, cmap='viridis', s=50)
plt.title("Fuzzy C-Means Clustering (2 Clusters)")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.grid(True)
plt.show()
```
