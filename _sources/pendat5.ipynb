###  3. Ambil contoh threshold split (misalnya):

* Threshold = 4.5 cm

Maka kita akan membagi:

* Group Left (≤ 4.5 cm) : Data No 13-10 -> 7 data, semua Rosa damascena
* Group Right (> 4.5 cm) : 8 data, semua Rosa galica

###  4. Hitung Entropy Awal (sebelum split)

$$
Entropy(D) = -p_1 \log_2(p_1) - p_2 \log_2(p_2)
$$

$$
p_1 = \frac{8}{15},\quad p_2 = \frac{7}{15}
$$

$$
Entropy(D) = -\frac{8}{15} \log_2 \frac{8}{15} - \frac{7}{15} \log_2 \frac{7}{15} \approx 0.996
$$

###  5. Entropy setelah split

Group Left (7 damasca): entropy = 0

* 7 data → 7 Rosa damascena, 0 Rosa gallica

Group Right (8 galica): entrpy = 0


Information Gain:

$$
IG = Entropy(D) 0.996 - \left( \frac{7}{15} \cdot 0 + \frac{8}{15} \cdot 0 \right) = 0.996
$$

###  6. Gini Index

$$
Gini(D) = 1 - \left( \frac{8}{15} \right)^2 - \left( \frac{7}{15} \right)^2 = 1 - \frac{64}{225} - \frac{49}{225} = 1 - \frac{113}{225} \approx 0.497
$$


Group Left (≤ 4.5 cm):

$$
  Gini_{left} = 1 - (0)^2 - (1)^2 = 1 - 0 - 1 = 0
$$

Group Right (0):

$$
Gini_{left} = 1 - (1)^2 - (0)^2 = 1 - 1 - 0 = 0
$$

Gini Split:

$$
Gini_{split} = \frac{7}{15} \cdot 0 + \frac{8}{15} \cdot 0 = 0
$$

## Visualisasi Decision Tree

Ini adalah code program untuk mencari decision tree sekaligus rules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree, export_text
from sklearn.preprocessing import LabelEncoder

# === Data Baru ===
data = {
    "Petal Length (cm)": [5.5, 5.6, 5.4, 5.2, 5.8, 4.2, 4.4, 4.3, 4.1, 4.5,
                          5.9, 5.3, 4.0, 5.7, 4.3],
    "Petal Width (cm)":  [2.1, 2.3, 2.2, 2.0, 2.4, 1.3, 1.4, 1.5, 1.2, 1.3,
                          2.5, 2.1, 1.1, 2.3, 1.2],
    "Kelas": [
        "Rosa gallica", "Rosa gallica", "Rosa gallica", "Rosa gallica", "Rosa gallica",
        "Rosa damascena", "Rosa damascena", "Rosa damascena", "Rosa damascena", "Rosa damascena",
        "Rosa gallica", "Rosa gallica", "Rosa damascena", "Rosa gallica", "Rosa damascena"
    ]
}

# Membuat DataFrame
df = pd.DataFrame(data)

# === Label Encoding ===
le = LabelEncoder()
df['Label'] = le.fit_transform(df['Kelas'])  # Rosa gallica = 0, Rosa damascena = 1

# === Fitur dan Target ===
X = df[["Petal Length (cm)", "Petal Width (cm)"]]
y = df["Label"]

# === Latih Decision Tree ===
clf = DecisionTreeClassifier(criterion="gini", max_depth=3, random_state=42)
clf.fit(X, y)

# === Tampilkan Aturan Decision Tree ===
print("\n=== Pohon Keputusan (Decision Tree) ===")
tree_rules = export_text(clf, feature_names=list(X.columns))
print(tree_rules)

# === Visualisasi Pohon Keputusan ===
plt.figure(figsize=(10, 3))
plot_tree(
    clf,
    feature_names=X.columns,
    class_names=le.classes_,
    filled=True,
    rounded=True,
    impurity=True
)
plt.title("Decision Tree: Rosa gallica vs Rosa damascena")
plt.tight_layout()
plt.show()
```

## Visualisasi Splitting

```python
uji_baru = pd.DataFrame({
    "Petal Length (cm)": [4.8, 5.1],
    "Petal Width (cm)":  [1.4, 2.2]
})
prediksi = clf.predict(uji_baru)
hasil    = le.inverse_transform(prediksi)

print("\n=== Prediksi 2 Data Baru ===")
for i, (pl, pw) in enumerate(zip(uji_baru["Petal Length (cm)"], uji_baru["Petal Width (cm)"])):
    print(f"Data {i+1}: Petal Length={pl}, Petal Width={pw} → Prediksi: {hasil[i]}")

# ---------- 7. Visualisasi Decision Boundary ----------
print("\n=== Visualisasi Decision Boundary ===")
plot_colors = "ry"   # r = gallica, y = damascena
plot_step   = 0.01

x_min, x_max = X["Petal Length (cm)"].min() - 0.2, X["Petal Length (cm)"].max() + 0.2
y_min, y_max = X["Petal Width (cm)"].min()  - 0.2, X["Petal Width (cm)"].max()  + 0.2

xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),
                     np.arange(y_min, y_max, plot_step))

grid   = np.c_[xx.ravel(), yy.ravel()]
grid_df = pd.DataFrame(grid, columns=["Petal Length (cm)", "Petal Width (cm)"])
Z      = clf.predict(grid_df).reshape(xx.shape)

plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, cmap=plt.cm.RdYlBu, alpha=0.6)

# Titik data training
for label, color in zip([0, 1], plot_colors):
    idx = np.where(y == label)
    plt.scatter(X.iloc[idx]["Petal Length (cm)"],
                X.iloc[idx]["Petal Width (cm)"],
                c=color, edgecolor="k",
                label=le.inverse_transform([label])[0])

# Titik data uji
plt.scatter(uji_baru["Petal Length (cm)"], uji_baru["Petal Width (cm)"],
            c="black", marker="X", s=100, label="Data Uji")

plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)")
plt.title("Decision Boundary (Petal Length vs Petal Width)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
